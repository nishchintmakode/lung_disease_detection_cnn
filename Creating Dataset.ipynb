```python
import pandas as pd
import os
import shutil
```

```python
FILE_PATH = "E:\Mini Project\Major Project\sample\sample_labels.csv"
IMAGES_PATH = "E:\Mini Project\Major Project\sample\images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR = "E:\Mini Project\Major Project\sample\dataset1"
```

```python
cnt = 0

for (i, row) in df.iterrows():
    if row["Finding Labels"] == "Nodule":
        filename = row["Image Index"]
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path, image_copy_path)
        print("Moving image ", cnt)
        cnt += 1
        
print(cnt)
```
